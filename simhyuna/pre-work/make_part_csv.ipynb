````python
### json파일 경로 지정 ###

label_dam_dir = '../data/Dataset/2.라벨링데이터/damage/'
label_part_dir = '../data/Dataset/2.라벨링데이터/damage_part/'

label_dam = os.listdir(label_dam_dir)
label_part = os.listdir(label_part_dir)
````

````python
### 원본 파일인 part_labeling.csv에서 train, test, val의 비율 알아보기 ###

part_labeling = pd.read_csv('part_labeling.csv')
part_labeling['dataset'].value_counts()
part_labeling['dataset'].value_counts(normalize=True)
````

````python
### (Part_json) 정보로 csv 만들기 ###

import warnings

warnings.filterwarnings(action='ignore')


### 컬럼명으로 df틀 만들기 ###
df = pd.DataFrame(columns = list(part_df.columns))


### 폴더의 json파일 하나씩 불러와서 filename 가져오기 ###
for i in range(100):
    with open(label_part_dir + f'{label_part[i]}') as file:
        part = json.load(file)
        file_name = part['images']['file_name']
        
        #print(len(part['annotations']))


        ### 파손 부위 해당 cell 값이 비었을 경우는 1 추가, 이미 값이 존재하면 (값 + 1)을 해당 cell에 입력 ###
        for j in range(len(part['annotations'])):
            part_name = part['annotations'][j]['part']
            if not part_name is None:
                if df.loc[df['img_id'] == file_name].empty:
                    df = df.append(pd.DataFrame([[file_name,1]], columns=['img_id',part_name]), ignore_index=True)
                else:
                    df.loc[df['img_id'] == file_name, part_name] = 1


### 총 파손 횟수인 total_anns 컬럼 및 컬럼값 생성 ###
df['total_anns'] = df.iloc[:,1:-3].sum(axis = 'columns')
df


### train:test:val = 8:1:1 비율로 만들기 ###
train = df.sample(frac = 0.8, random_state= 7)
train['dataset'] = 'train'
remain = df.drop(index = train.index)
test = remain.sample(frac = 0.5, random_state= 7)
test['dataset'] = 'test'
val = remain.drop(index= test.index)
val['dataset'] = 'val'


### train, test, val df 합친 후 정렬 ###
df = pd.concat([train, test, val], ignore_index=True)
df = df.sort_values(by = 'img_id')
df.to_csv("part_labeling.csv", index = False)
df
````
