```python
import warnings

warnings.filterwarnings(action='ignore')

df = pd.DataFrame(columns = list(part_df.columns))


for i in range(10):
    with open(label_part_dir + f'{label_part[i]}') as file:
        part = json.load(file)
        file_name = part['images']['file_name']
        
        #print(len(part['annotations']))
        
        for j in range(len(part['annotations'])):
            part_name = part['annotations'][j]['part']
            if not part_name is None:
                if df.loc[df['img_id'] == file_name].empty:
                    df = df.append(pd.DataFrame([[file_name,1]], columns=['img_id',part_name]), ignore_index=True)
                else:
                    df.loc[df['img_id'] == file_name, part_name] = 1


df
```
